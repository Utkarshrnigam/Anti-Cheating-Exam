```python
import os
import glob
import pandas as pd
import shutil
```

```python
side = []
frontal = []
target_side = "./final_dataset/side_face/"
target_frontal = "./final_dataset/frontal_face/"
paths = './final_dataset/face_dataset_RGB/'
s = 0
f = 0
for root,dirs,files in os.walk(paths):
    if len(dirs)>0:        
        for folder in dirs:
            for root,dirs,files in os.walk(paths+folder):
                print(folder)
                df = pd.read_csv(paths+folder+"/"+files[0],sep = '\t',header=None,dtype={0: object})
                data = df.values
                data = data[:,:2]
                for i in data:
                    i[0] = "frame_" + i[0] + "_face_rgb.png"
                
                for i in data:
                    if i[1]>=22 or i[1]<=-22:
                        s = s+1
                        shutil.copy(paths+folder+"/"+i[0], target_side+"/"+str(s)+".png")
                        side.append(i[0])
                    else:
                        f = f+1
                        shutil.copy(paths+folder+"/"+i[0], target_frontal+"/"+str(f)+".png")
                        frontal.append(i[0])
            
```

```python
import numpy as np
import cv2
from mtcnn.mtcnn import MTCNN
import matplotlib.pyplot as plt
    
```

```python
detector = MTCNN()
```

```python
path_side = "./final_dataset/side_face/"
path_frontal = "./final_dataset/frontal_face/"
target_cropped_frontal = "./final_dataset/cropped_frontal/"
target_cropped_side = "./final_dataset/cropped_side/"
j=0
for root,dirs,files in os.walk(path_frontal):
    for file in files:
        j = j+1
        if j%1000 == 0:
            print(j)
        
        full_file = path_frontal+file
        img = cv2.imread(full_file)
#         img = cv2.cvtColor(img,cv2.COLOR_BGR2RGB)
        faces = detector.detect_faces(img)
        for i in range(len(faces)):
            bb = faces[i]['box']
            x, y, w, h = bb
            f_img = img[y:y+h,x:x+w]
            cv2.imwrite(target_cropped_frontal+file,f_img)

        


        
```
